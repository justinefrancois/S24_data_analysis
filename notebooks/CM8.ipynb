```python
def mean_list_diff(a,b):
    mean1 = sum(a)/len(a)
    mean1 = sum(b)/len(b)
    mean_diff = mean1 - mean2
    return mean_diff
```

```python
boot_arr= []
for i in range (n):
    boot_v1 = choices(v1, k=len(v1))
    boot_v2 = choices(v2, k=len(v2))
    boot_arr.append (mean_list_diff(boot_v2, boot_v1))
```

```python
alpha = 0.01
```

```python
sorted_boot_arr = sorted(boot_arr)
```

```python
left_side = int(len(sorted_boot_arr) * alpha/2)
right_side = int(len(sorted_boot_arr) * (1 - alpha/2))
```

```python
ls = sorted_boot_arr[left_side]
rs = sorted_boot_arr[right_side]

print ("[", round(ls, 5), ",", round(rs, 5), "]")
```

```python
def calc_ci(a):
    left_side = int(len(sorted_boot_arr) * alpha/2)
    right_side = int(len(sorted_boot_arr) * (1 - alpha/2))
    ls = sorted_boot_arr[left_side]
    rs = sorted_boot_arr[right_side]
    ci = [ls, rs]
    return ci
```
